```python
import pandas as pd

df = pd.read_csv("HR.csv")
print(df.shape)
print(df.head())
print(df.info())
```

```python
print(df["Attrition"].value_counts())
print(df.groupby("Attrition")["MonthlyIncome"].mean())
```

```python
print(df.groupby("Attrition")["OverTime"].value_counts())
```

```python
print(df.groupby("Department")["Attrition"].value_counts(normalize=True))
```

```python
print(df.groupby("Attrition")["JobSatisfaction"].mean())
print(df.groupby("Attrition")["YearsAtCompany"].mean())
```

```python
import matplotlib.pyplot as plt

df.groupby("Attrition")["OverTime"].value_counts(normalize=True).unstack().plot(kind="bar")
plt.title("Overtime Rate by Attrition Status")
plt.ylabel("Proportion")
plt.savefig("overtime_chart.png")
plt.show()
```

```python
# Income comparison
df.groupby("Attrition")["MonthlyIncome"].mean().plot(kind="bar", color=["steelblue","orange"])
plt.title("Average Monthly Income by Attrition Status")
plt.ylabel("Monthly Income ($)")
plt.savefig("Monthly_Income.png")
plt.show()

# Department attrition rate
df.groupby("Department")["Attrition"].value_counts(normalize=True).unstack()["Yes"].plot(kind="bar", color="orange")
plt.title("Attrition Rate by Department")
plt.ylabel("Attrition Rate")
plt.savefig("Attrition_Rate.png")
plt.show()
```
